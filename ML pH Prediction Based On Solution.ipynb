### Use Of Scikit_Learn

```python
# '''so now we  use Scikit learn because it is ML library and we can perform our tasks on it'''b
```

```python
# now we will target our features and labels
#1st we will assign the feature column
x=df[['Conductivity_mS','Temperature_C']]
# 2nd we target the label and predict the output result
y=df[['pH']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
### cal the model and train the model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# model.predict([[74.524892,5.803020]])
```

```python
# import matricss to Evalutate the model we have just trained
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score,root_mean_squared_error,mean_squared_error
```

## *Results*

```python
y_pred=model.predict(x_test)
print(f'MAE:{mean_absolute_error(y_test,y_pred)}')
print(f'MAPE:{mean_absolute_percentage_error(y_test,y_pred)}')
print(f'R2:{r2_score(y_test,y_pred)}')
print(f'root_mean_squared_error:{root_mean_squared_error(y_test,y_pred)}')
```
